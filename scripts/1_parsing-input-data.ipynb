```python
### RAW TEXT FROM CSEL
CSEL_raw_text_list = []
for dirpath, dirs, files in os.walk("../data_input/CSEL_data"):
    for file_path in files:
        if file_path.endswith(".xml"):
            if not file_path.startswith("__"):
                with open(dirpath + "/" + file_path) as file:
                    soup = BeautifulSoup(file, 'html.parser')
                    file_name = os.path.basename(file.name)
                try:
                    author = str(soup.monogr.author.get_text())
                except:
                    author = "unknown"
                try:
                    title = str(soup.monogr.title.get_text())
                except:
                    title = "unknown"
                try:
                    volume = str(soup.monogr.biblScope.get_text())
                except:
                    volume = "unknown"
                raw_text = soup.body.get_text()
                CSEL_raw_text_list.append([file_name, author, title, volume, raw_text])
CSEL_raw_df = pd.DataFrame(CSEL_raw_text_list, columns=["file_name", "author", "title", "volume","raw_text"])
with open('data/CSEL_raw.csv', 'a') as f: 
    CSEL_raw_df.to_csv(f, header=False)
```
